```python
import pandas as pd

# Load the data
file_path = 'Cleaned_SIPRI_Military_Expenditure-Constant_2022.csv'
df1 = pd.read_csv(file_path)

# Display the first few rows to understand the structure
df1.head()
```

```python
# Data Preprocessing
# Transpose the dataframe to make it easier to work with years as time series data for each country

df1_long = df1.melt(id_vars=['Country'], var_name='Year', value_name='Expenditure_Constant_2022')
df1_long['Year'] = pd.to_numeric(df1_long['Year'])
df1_long.dropna(inplace=True)

# Display the first few rows of the long format data
df1_long.head()
```

```python
# Output the cleaned DataFrame with country-specific military expenditure data to a CSV file
output_path = "Cleaned_Long_SIPRI_Military_Expenditure-Constant_2022.csv"
df1_long.to_csv(output_path, index=False)

output_path
```

```python
# Merge four csv into one

import pandas as pd

# Load the uploaded CSV files
file_paths = [
    'Cleaned_Long_SIPRI_Military_Expenditure-Share_of_Govt_spending.csv',
    'Cleaned_Long_SIPRI_Military_Expenditure-Share_of_GDP.csv',
    'Cleaned_Long_SIPRI_Military_Expenditure-Per_Capita.csv',
    'Cleaned_Long_SIPRI_Military_Expenditure-Constant_2022.csv'
]

# Read each file into a DataFrame
govt_spending_df = pd.read_csv(file_paths[0])
gdp_share_df = pd.read_csv(file_paths[1])
per_capita_df = pd.read_csv(file_paths[2])
constant_2022_df = pd.read_csv(file_paths[3])

# Merge the datasets on common columns 'Country' and 'Year'
merged_df = govt_spending_df.merge(gdp_share_df, on=['Country', 'Year'], how='outer')
merged_df = merged_df.merge(per_capita_df, on=['Country', 'Year'], how='outer')
merged_df = merged_df.merge(constant_2022_df, on=['Country', 'Year'], how='outer')

# Rename the columns
merged_df.columns = [
    'Country', 'Year',
    'Expenditure-Share_of_Govt_spending', 
    'Expenditure-Share_of_GDP',
    'Expenditure_Per_Capita',
    'Expenditure_Constant_2022'
]

# Set 'Index' column as the dataframe index
merged_df.reset_index(inplace=True, drop=True)
merged_df.index.name = 'Index'

# Display the final merged DataFrame
merged_df
```

```python
# Output the merged DataFrame to a CSV file
output_path = 'Cleaned_Long_SIPRI_Military_Expenditure_Merged.csv'
merged_df.to_csv(output_path, index=False)

output_path
```
